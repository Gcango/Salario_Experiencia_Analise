##Uma visão sobre o dataset do Kaggle, onde constam dados sobre salarios referente dos funcionários. Neste projecto foi importante analisar o Tempo de Experiencia vs O Salário de um determinado Funcionário.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
dados = pd.read_csv("/content/Salary_Data.csv")
```

```python
dados.head()
```

```python
print(" Dimensão da Base de dados, total de linhas e colunas", dados.shape)
```

```python
dados.describe()
```

```python
dados.isnull().sum()
```

```python
sns.heatmap(dados.isnull(), cbar=False);
```

```python
dados.duplicated().any()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Analise de Renda')
sns.kdeplot(dados['Salary'], color='green');
```

```python
plt.figure(figsize=(10, 6))
plt.title('Analise de Renda')
sns.kdeplot(dados['YearsExperience'], color='blue');
```

```python
#Vendo a Distribuição
plt.figure(figsize=(8, 9))
sns.boxplot(dados['Salary']);
```

```python
#Vendo a Distribuição
plt.figure(figsize=(8, 9))
sns.boxplot(dados['YearsExperience']);
```

```python
# Distribuiçã
plt.figure( figsize=(10, 5) )
sns.scatterplot( data=dados, x='Salary', y='YearsExperience');
```

```python
# Distribuiçã
plt.figure( figsize=(10, 5) )
sns.regplot( data=dados, x='Salary', y='YearsExperience');
```

```python
# Correlação
Correlacao = dados.corr('pearson')
plt.figure( figsize=(10, 5) )
sns.heatmap( Correlacao, annot=True, cbar=True );
```

```python
# Correlação
Correlacao = dados.corr('spearman')
plt.figure( figsize=(10, 5) )
sns.heatmap( Correlacao, annot=True, cbar=True );
```
